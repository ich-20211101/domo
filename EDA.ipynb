```python
from datetime import datetime

# UTC time
current_time = datetime.now()
print(current_time)
print("Current time:", current_time.strftime("%H:%M:%S"))

# Pacific time
pacific_time = current_time - pd.Timedelta(hours=8)
print(pacific_time)
print("Pacific time:", pacific_time.strftime("%H:%M:%S"))
```

```python
# Calculate the difference between Plus & MaxPlus
# and store it in a new cloumn "DifferenceOfPulse"

data['DifferenceOfPulse'] = data['Maxpulse'] - data['Pulse']
data.head()
```

```python
# Handling missing values
# Different ways of handling missing valuse:
# Mean / Median / Mode / Drop those rows

# Mean
data['Calories'] = data['Calories'].fillna(data['Calories'].mean())

print('Mean of Maxpulse is:', data['Maxpulse'].mean())
print('Median of Maxpulse is:', data['Maxpulse'].median())
print('Mode of Maxpulse is:', data['Maxpulse'].mode()[0])

data['Calories'].mean()
```
