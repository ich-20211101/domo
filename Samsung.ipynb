```python
from datetime import date

data['Date'] = pd.to_datetime(data['Date'])
data['Data_Age'] = date.today().year - data['Date'].dt.year

data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = data.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    print(col)
    print('Skew: ', round(data[col].skew(), 2))

    plt.figure(figsize = (15, 4))

    plt.subplot(1, 2, 1)
    data[col].hist(grid = False)
    plt.ylabel('Count')

    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])

    plt.show()
```

```python
import numpy as np

def log_transform(data, col):
    for colname in col:
        if (data[colname] == 1.0).all():
          data[colname + '_log'] = np.log(data[colname] + 1)
        else:
          data[colname + '_log'] = np.log(data[colname])
    data.info()

log_transform(data, ['High', 'Low'])
sns.distplot(data['High_log'], axlabel='High_log')
```

```python
plt.figure(figsize = (13, 17))
sns.pairplot(data = data.drop(['High', 'Low'], axis = 1))
plt.show()
```

```python
plt.figure(figsize = (12, 7))
sns.heatmap(data.drop(['High', 'Low'], axis = 1).corr(), annot = True, vmin = -1, vmax = 1)
plt.show()
```
